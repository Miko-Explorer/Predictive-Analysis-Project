# Logistics Regression Model (Baseline Model)
---
**Definition of the model:**
* This model is a classification model used to predict categories.
* It is commonly used for binary classification, such as predicting high risk (1) or low risk (0).
* It works by finding a relationship between the input and the probability of a certain class.

**Complete syntax and description (Pyhton/ scikit-learn):**
```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```
*from sklearn.linear_model import LogisticRegression*
* Imports the logistics regression model from scikit-learn

*model = LogisticRegression( )*
* Creates the model with default settings.
* You can also add parameters here like max_iter = 1000

*model.fit(X-train, y_train)*
* Trains the model
* **X_train** = is the training dataset with input features (like age, income, salary).
* **y_train** =  is the target variable (0 or 1) or the target output or prediction (what you want to predict like "yes"/"no" or"high"/"low").
* **.fit( )** = The model "learns" from the data - finds patterns and relationships.

*predictions = model.predict(X_test)*
* Uses tha trained model to make predictions on new data.
* **X_test** = new data the model hasnt seen before.
* **.predict( )** = uses what it learned to make predictions.
* **predictions** = The model's guessess (like "low risk" or "high risk").

**Necessary inputs:**
* **X values (features):** numerical columns used to make predictions.
* **y values (target):** a binary label such as 0 or 1.
* **Clean data:** no missing values, properly encoded.

**Things to tweak:**
* **max_iter:** number of training cycles. Increase if the model does not converge.
* **C (Regularization strength):** controls overfitting.
* **Feature selection:** choose only important features.
* **Scaling:** use StandadrdScalar for better performance.

**Expected Output:**
* A predicted category such as 0 = low risk or 1 = high risk.
* You can also get probability scores using ***.predict_proba()***.

**Different Parameters for LogisticRegression( ):**
* **penalty:** controls how strictly the model should follow the rules (The rule enforcer).
1. **penalty='l2'** - (default) gentle rule enforcement - "Try to keep weights small"
2. **penalty='l1'** - struct rule enforcement - "Forse some weights to be exactly zero"
3. **penalty='elasticnet'** - mix of both strict and gentle.
4. **penalty='none'** - no rules - "Do whatever works".
* **C:** controls how flexible or rigid the model should be or how closely the model should follow a given guide (The flexibility knob).
1. **C=0.01** - very rigid model - "Stick exactly to the rule".
2. **C=100** - very flexible model - "Break rules if it helps predict better".
3. **C=1.0** - Balanced approach.
* **solver:** chooses the math method to find the best solution (in leymans term: different routes same destination).
1. **solver='lbfgs'** - (default) good for most problems, works l2 penalty.
2. **solver='liblinear'** - good for small datasets, works with l1 and l2 penalty.
3. **solver='saga'** - fast for large datasets, works with all penalties.
4. **solver='newton-cg'** - good for medium datasets.
* **max_iter:** maximum number of attempts to find the best solution (in leymans term: how many time you'll try to solve).
1. **max_iter=100, max_iter=1000, max_iter=10000** - these are common values for max_iter.
* **random_state:** makes results reproducible (in leymans term: same numbers to get the same result every time you run your code).
1. **random_state=42** - this is a common practice (but can be any number).
* **class_weight:** handles imbalanced datasets (The fairness adjuster).
1. **class_weight='None'** - (default) treat all classes equally.
2. **class_weight='balanced'** - automatically adjust for class imbalance.
3. **class_weight='dictionary'** - manually specify weights: {0:1, 1:10} (class  is 10x more important).
* **fit_intercept:** whether to include a starting point/baseline in the model.
1. **fit_intercept=True** - (default) include a baseline - usually better.
2. **fit_intercept=False** - force the model through origin -rarely used.

# Random Forest Classifier (Advanaced Model)
---
**Definition of the model:**
* Random Forest is an advanced classificatio model that uses many decision trees.
* Each tree makes its own prediction, and the forest picks the majority vote.
* It is good for complex datasets and usually more accurate than simple models.

**Complete syntax and description (Pyhton/ scikit-learn):**
```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```
*from sklearn.ensemble import RandomForestClassifier*
* Imports the random forest classifier model.

*model = RandomForestClassifier()*
* Creates the model with default settings.
* You can set parameters like n_estimators = 100

*model.fit(X-train, y_train)*
* Trains the model using many decision trees.
* **X_train**= your inpit features (all the information you have).
* **y_train**= what you want to predict like "high"/"low".
* **.fit( )**= the forest "grows" - each three learns patterns from the data.

*predictions = model.predict(X_test)*
* Predicts the category for new data.
* **X_test**= new data the forest hasn't seen.
* **.predict( )** = all trees vote on the answer, majority wins.
* **predictions** = the forest's collective decision.  

**Necessary inputs:**
* **X values (features):** all important variables in the dataset.
* **y values (target):** binary labels (0 or 1).
* **Clean data:** no missing values, properly encoded.

**Things to tweak:**
* **n_estimators:** number of tress. More trees can improve accuracy.
* **max_depth:** how deep the trees can grow. Controls overfitting.
* **min_samples_split/mim_samples_leaf:** controls how trees split.
* **Random state:** helps keep results consistent.

**Expected Output:**
* A predicted class such as 0 = low risk or 1 = high risk.
* You can also get: feature important scores, probability scores, more acccurated predictions comapred to baseline.

**Different Parameters for RandomForestClassifier( ):**
* **n_estimators=100** - number of trees (more trees = more accurate results but slower).
* **max_depth=None** - how complex each tree can be (None = grow fully).
* **random_state=42** - for same results every time.
* **min_samples_split=2** - minimum sample to split a node.
* **min_samples_leaf=1** - minimum sample in a leaf node.

# Evaluate Performance
---
**Complete Synatx for Evaluation of Performance for each model:**
```python
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score, confusion_matrix)
from sklearn.metrics import classification_report

accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)  
f1 = f1_score(y_true, y_pred)

#Logistics Regression Report:
print(classification_report(y_test, logistic_pred))
#Random Forest Classifier Report:
print(classification_report(y_test, rf_pred))
```
**Accuracy:**
* Percentage of correct predictions.
* **Formula:** (Correct Prediction)/(Total Predictions).
* **When to use:** When classes are balanced.
* **Whatch out:** Can be misleading with imbalanced data.
* **Syntax:** *accuracy = accuracy_score(y_true, y_pred)*

**Precision:**
* Of the "positive" predictions, how many were actually positive.
* Ex: "When I say yes", how often I am right.
* **Formula:** True Positives/(True positives + False positives).
* **Good for:** when false alarms are expensive
* **Syntax:** *precision = precision_score(y_true, y_pred)*

**Recall Sensitivity:**
* Of all actual positives, how many did we catch.
* Ex: "Of all sick people, how many did we identify"
* **Formula:** True Positives/(True positives + False negatives).
* **Good for:** When missing positives are dangerous.  
* **Syntax:** *recall = recall_score(y_true, y_pred)*

**F1-Score**
* balance between precision and recall
* Ex: "Harmonic mean" - balances tow important things.
* **Formula:** 2 x (precision x recall)/(precision + recall).
* **Good for:** single number that balances both concerns.  
* **Syntax:** *f1 = f1_score(y_true, y_pred)*

**Parameters per function mentioned above:**
* **y_true** = actual answers (y_test from train_test_split( )).
* **y_pred** = model predictions (predictions = model.predict(X_test)).

**Classiification Report For each Model:**
* **Logistics Regression** = *classification_report(y_test, logistic_pred)*
* **Random Forest Classifier** = *classification_report(y_test, rf_pred)*

**2.) Import all necessary libraries:**

````python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
````

**3.) Load dataset and preview of the dataset:**

````python
# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/2nd YR /1st Semester/1st Sem - Major Courses/DSA#4153/Finals & Prelims Activities/GROUP#3 - Finals Project/Dataset/GROUP#3_2DSA2_FINAL_DATASET.csv")

# Preview of dataset:
print(df)
````

**4.) Prepare the data and the dataset:**

````python
print("="*70)
print("DISEASE OUTBREAK RISK PREDICTION SYSTEM")
print("="*70)

# SECTION 1: Prepare the data

# Sort by Region, Disease, and Morbidity Week (MW)
df = df.sort_values(['Region', 'Disease', 'MW'])

# Create target variable: Will NEXT WEEK have high cases?
# Shift cases forward by 1 week for each Region-Disease combination
df['next_week_cases'] = df.groupby(['Region', 'Disease'])['No. of Cases'].shift(-1)

# Calculate average cases (use current week's average for fair comparison)
average_cases = df['No. of Cases'].mean()

# Create target: 1 if NEXT WEEK has above-average cases, 0 otherwise
df['risk_prediction'] = np.where(df['next_week_cases'] > average_cases, 1, 0)

# Remove the last week for each group (no future data to predict)
df = df.dropna(subset=['risk_prediction', 'next_week_cases'])

print(f"Dataset loaded: {df.shape[0]} records")
print(f"Average cases per location: {average_cases:.1f}")
print(f"High risk locations: {sum(df['risk_prediction']==1)}")
print(f"Low risk locations: {sum(df['risk_prediction']==0)}")
print()
````

**5.) Prepare data for Random Forest Classifier:**

````python
# Convert text columns to numbers using one-hot encoding
df_encoded = pd.get_dummies(df, columns=['Region', 'PHI', 'Barangay', 'Disease'], drop_first=True)

# Prepare features and target for Random Forest
rf_features = [col for col in df_encoded.columns if col != 'risk_prediction']
X_rf = df_encoded[rf_features]
y_rf = df_encoded['risk_prediction']

print(f"Random Forest will use {len(rf_features)} features")
print()
````

**6.) Logistics Regression Model:**

````python
# Uses only number of cases and MW as features
print("-"*60)
print("LOGISTIC REGRESSION MODEL")
print("-"*60)
print("Using features: No. of Cases and MW")

# Select only two features for Logistic Regression
log_features = ['No. of Cases', 'MW']
X_log = df[log_features]
y_log = df['risk_prediction']

# Split data into training and testing sets
X_log_train, X_log_test, y_log_train, y_log_test = train_test_split(
    X_log, y_log, test_size=0.3, random_state=42, stratify=y_log
)

# Create and train Logistic Regression model
log_model = LogisticRegression(random_state=42, max_iter=1000, class_weight= 'balanced')
log_model.fit(X_log_train, y_log_train)

# Make predictions on test data
log_predictions = log_model.predict(X_log_test)

# Calculate performance metrics
log_accuracy = accuracy_score(y_log_test, log_predictions)
log_precision = precision_score(y_log_test, log_predictions, zero_division=0)
log_recall = recall_score(y_log_test, log_predictions, zero_division=0)
log_f1 = f1_score(y_log_test, log_predictions, zero_division=0)

# Display first 20 predictions
print("\nPredictions (First 20 test samples):")
print("-"*60)
print(f"{'Region':<8} {'Cases':<8} {'MW':<4} {'Actual':<8} {'Predicted':<10}")
print("-"*60)

test_indices = list(y_log_test.index)[:18000]
for i, idx in enumerate(test_indices):
    region = df.loc[idx, 'Region']
    cases = df.loc[idx, 'No. of Cases']
    mw = df.loc[idx, 'MW']

    actual = "HIGH" if y_log_test.iloc[i] == 1 else "LOW"
    predicted = "HIGH" if log_predictions[i] == 1 else "LOW"

    print(f"{region:<8} {cases:<8} {mw:<4} {actual:<8} {predicted:<10}")

# Display model performance
print("-"*60)
print("\nModel Performance:")
print(f"Accuracy:  {log_accuracy:.2%}")
print(f"Precision: {log_precision:.2%}")
print(f"Recall:    {log_recall:.2%}")
print(f"F1 Score:  {log_f1:.2%}")
print()
````

**7.) Random Forest Classifier Model:**

````python
# Uses all available features including encoded categorical data
print("-"*60)
print("RANDOM FOREST MODEL")
print("-"*60)
print(f"Using all {len(rf_features)} available features")

# Split data for Random Forest
# Updates: updated random_state from 99 to 42
X_rf_train, X_rf_test, y_rf_train, y_rf_test = train_test_split(
    X_rf, y_rf, test_size=0.3, random_state=42, stratify=y_rf
)

# Create and train Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42,  class_weight= 'balanced')
rf_model.fit(X_rf_train, y_rf_train)

# Make predictions
rf_predictions = rf_model.predict(X_rf_test)

# Calculate performance metrics
rf_accuracy = accuracy_score(y_rf_test, rf_predictions)
rf_precision = precision_score(y_rf_test, rf_predictions, zero_division=0)
rf_recall = recall_score(y_rf_test, rf_predictions, zero_division=0)
rf_f1 = f1_score(y_rf_test, rf_predictions, zero_division=0)

# Get the test indices (these are indices from the original df)
test_indices = X_rf_test.index

# Match the approach from Logistic Regression section
print("\nPredictions (First 20 test samples):")
print("-"*60)
print(f"{'Region':<8} {'Cases':<8} {'MW':<4} {'Actual':<8} {'Predicted':<10}")
print("-"*60)

# Get indices from y_rf_test and match to original df
test_indices = list(y_rf_test.index)[:18000]

for i, idx in enumerate(test_indices):
    # Use .loc with the original index (this worked for Logistic Regression)
    try:
        region = df.loc[idx, 'Region']
        cases = df.loc[idx, 'No. of Cases']
        mw = df.loc[idx, 'MW']

        actual = "HIGH" if y_rf_test.iloc[i] == 1 else "LOW"
        predicted = "HIGH" if rf_predictions[i] == 1 else "LOW"

        print(f"{region:<8} {cases:<8} {mw:<4} {actual:<8} {predicted:<10}")
    except KeyError:
        # If index doesn't exist in df, skip detailed display
        actual = "HIGH" if y_rf_test.iloc[i] == 1 else "LOW"
        predicted = "HIGH" if rf_predictions[i] == 1 else "LOW"
        print(f"Sample {i+1:<2}   {actual:<8} {predicted:<10}")

# Display model performance
print("-"*60)
print("\nModel Performance:")
print(f"Accuracy:  {rf_accuracy:.2%}")
print(f"Precision: {rf_precision:.2%}")
print(f"Recall:    {rf_recall:.2%}")
print(f"F1 Score:  {rf_f1:.2%}")
print()
````
